```python
#!/usr/bin/env python
# coding: utf-8
import pandas as pd
train_data = pd.read_csv('train_data_public.csv')
train_data
```

```python
# 将BIO_anno转换为 list形式
train_data['BIO_anno'] = train_data['BIO_anno'].apply(lambda x: x.split(' '))
# 拼接text与BIO_anno
train_data['training_data'] = train_data.apply(lambda row: (list(row['text']), row['BIO_anno']), axis=1)
train_data
```

```python
train_data['training_data']
```

```python
test_data = pd.read_csv('test_public.csv')
test_data
```

```python
test_data['testing_data'] = test_data.apply(lambda row: (list(row['text'])), axis=1)
test_data
```

```python
training_data_txt = []
testing_data_txt = []

for i in range(len(train_data)):
    training_data_txt.append(train_data.iloc[i]['training_data'])

for i in range(len(test_data)):
    testing_data_txt.append(test_data.iloc[i]['testing_data'])
```

```python
testing_data_txt
```

```python
training_data = training_data_txt
word_to_label = {}
for sentence, tags in training_data:
    for i, word in enumerate(sentence):
        if word not in word_to_label:
            word_to_label[word] = tags[i]
```

```python
f = open("word_to_label.txt", 'w', encoding='utf-8')
for word, label in word_to_label.items():
        f.write(word + ' ' + label + '\n')
```
